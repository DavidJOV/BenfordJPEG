```python
from bs4 import *
import requests
import os
import urllib



def folder_create(images,destination):
    try:
        folder_name = os.path.join(destination)
        os.mkdir(folder_name)
    except:
        print("Verzeichnis ist bereits vorhanden! " + folder_name)
        folder_create()

    download_images(images, folder_name)


def download_images(images, folder_name):
    count = 0
    print(f"Es wurden {len(images)} Bilder gefunden :"+destination)
    if len(images) != 0:
        for i, image in enumerate(images):
            #Maximal 100 Bilder herunterladen pro durchlauf
            if i == 101:
                break;
            try:
                image_link = image["data-srcset"]
            except:
                try:
                    image_link = image["data-src"]
                except:
                    try:
                        image_link = image["data-fallback-src"]
                    except:
                        try:
                            image_link = image["src"]
                        except:
                            pass

            try:
                r = requests.get(image_link,timeout=(10, 10)).content
                try:
                    r = str(r, 'utf8')

                except UnicodeDecodeError:
                    with open(f"{folder_name}/images{i + 1}.jpg", "wb+") as f:
                        f.write(r)
                    count += 1
            except:
                pass

        if count == len(images):
            print("Dowload bendedet!")
        else:
            print(f"Es wurden {count} von {len(images)} Bildern heruntergeladen: "+destination)


def main(url,destination):
    r = requests.get(url,timeout=(10, 10))
    soup = BeautifulSoup(r.text, 'html.parser')
    images = soup.findAll("img")
    if len(images)==0:
        main(url,destination)
    else:
        folder_create(images,destination)


# ryanairDestinations = [ "Barcelona", "Belfast", "Billund", "Bordeaux", "Bratislava", "Bremen", "Brindisi", "Casablanca",
#                         "Chania", "Cork", "Deauville", "Dubrovnik", "Frankfurt", "Glasgow", "Haugesund",
#                         "Jerez de la Frontera", "Karlsruhe / Baden-Baden", "Lisbon", "London", "Lourdes Tarbes",
#                         "Milan", "Murcia", "Nuremberg", "Odesa", "Poitiers", "Preveza/Lefkada", "Rome", "Stockholm",
#                         "Tangier", "Tenerife", "Tours", "Varna", "Vilnius", "Łódź", "Aalborg", "Aarhus", "Aberdeen",
#                         "Agadir", "Alghero", "Alicante", "Almería", "Amman", "Amsterdam", "Ancona", "Aqaba", "Athens",
#                         "Banja Luka", "Bari", "Basel", "Beirut", "Bergerac", "Berlin", "Biarritz", "Bilbao",
#                        "Birmingham", "Bodrum", "Bologna", "Bournemouth", "Brest", "Bristol", "Brive-la-Gaillarde",
#                        "Brno", "Brussels", "Bucharest", "Budapest", "Burgas", "Bydgoszcz", "Béziers", "Cagliari",
#                        "Carcassonne", "Cardiff", "Castellón", "Catania", "Charleroi", "Châlons", "Ciudad Real",
#                        "Clermont-Ferrand", "Cluj", "Cologne / Bonn", "Comiso", "Copenhagen", "Corfu", "Craiova",
#                        "Crotone", "Cuneo", "Dalaman", "Derry", "Dole", "Dortmund", "Dresden", "Dublin", "Düsseldorf",
#                        "East Midlands", "Edinburgh", "Eilat", "Eindhoven", "Erfurt", "Essaouira", "Exeter", "Faro",
#                        "Fez", "Figari", "Freiburg", "Friedrichshafen", "Fuerteventura", "Galway", "Gdańsk", "Genoa",
#                        "Girona", "Glasgow", "Gothenburg", "Grenoble", "Gyumri", "Hahn", "Hamburg", "Hanover",
#                        "Helsinki", "Heraklion", "Ibiza", "Inverness", "Kalamata", "Katowice", "Kaunas", "Kavala",
#                        "Kefalonia", "Kerry", "Kharkiv", "Kherson", "Kiev", "Knock", "Kos", "Košice", "Kraków",
#                        "Kutaisi", "La Rochelle", "Lamezia Terme", "Lanzarote", "Lappeenranta", "Larnaca", "Las Palmas",
#                        "Leeds / Bradford", "Leipzig", "Lille", "Limoges", "Liverpool",
#                        "London", "Lorient", "Lublin", "Luxembourg City", "Lviv", "Lyon", "Maastricht", "Madrid", "Magdeburg",
#                        "Malmö", "Manchester", "Marrakesh", "Marseille", "Memmingen", "Menorca", "Milan", "Montpellier", "Mulhouse",
#                        "Munich", "Murcia", "Mykonos", "Málaga", "Münster", "Nador", "Nantes", "Naples", "Newcastle upon Tyne", "Newquay",
#                        "Nice", "Niš", "Nîmes", "Olbia", "Olsztyn", "Oradea", "Osijek", "Oslo", "Ostend", "Ostrava", "Ouarzazate", "Oujda",
#                        "Ovda", "Paderborn", "Palanga", "Palermo", "Palma de Mallorca", "Paphos", "Pardubice", "Parma", "Perpignan",
#                        "Perugia", "Pescara", "Pisa", "Plovdiv", "Podgorica", "Ponta Delgada", "Porto", "Poznań", "Prague", "Pula", "Rabat",
#                         "Reus", "Rhodes", "Riga", "Rijeka", "Rimini", "Rodez", "Rome", "Rygge", "Rzeszów", "Saint-Malo Dinard", "Saint-Étienne",
#                         "Salzburg", "Santander", "Santiago de Compostela", "Santorini", "Seville", "Shannon", "Sofia", "Split", "Stockholm", "Strasbourg",
#                         "Stuttgart", "Szczecin", "Tallinn", "Tampere", "Tbilisi", "Teesside", "Tel Aviv", "Tenerife", "Terceira Island", "Thessaloniki",
#                         "Tillé", "Timișoara", "Toulouse", "Trapani", "Trieste", "Turin", "Turku", "Tétouan", "Valencia", "Valladolid", "Valletta Luqa",
#                         "Venice", "Verona", "Vienna", "Vigo", "Vitoria-Gasteiz", "Växjö", "Warsaw", "Waterford", "Wrocław", "Zadar", "Zagreb", "Zakynthos",
#                         "Zaragoza", "Yerevan"]

# for destination in ryanairDestinations:
#     try:
#         main("http://www.airbnb.de/s/"+urllib.parse.quote(str(destination),safe="")+"/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_dates%5B%5D=august&flexible_trip_dates%5B%5D=september&flexible_trip_lengths%5B%5D=weekend_trip&date_picker_type=calendar&checkin=2021-08-13&checkout=2021-08-15&source=structured_search_input_header&search_type=filter_change",destination)
#     except:
#         print(destination+ " konnte nicht abgerufen werden!")
```
