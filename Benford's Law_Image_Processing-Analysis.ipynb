```python
import pandas as pd
import numpy as np
import math
import cv2
import os
```

```python
def firstDigit(number):
    digits = (int)(math.log10(number)) 
    number = (int)(number / pow(10, digits)) 
    return number
def graph(image_name,flag,directoryName,filename):
    print(directory,filename)
    try:
        img = cv2.imread(image_name,0)
        norm_img = np.zeros((800,800))
        final_img = cv2.normalize(img,  norm_img, 0, 255, cv2.NORM_MINMAX)
        cv2.imwrite(directoryName+"_"+filename+".jpg", final_img)
        img = cv2.imread("./"+directoryName+"_"+filename+".jpg",0)
        imf = np.float32(img)/255.0
        dst = cv2.dct(imf)
        #CleanUp
        if os.path.exists("./"+directoryName+"_"+filename+".jpg"):
            os.remove("./"+directoryName+"_"+filename+".jpg")
        else:
            print("The file does not exist")

        col = dst.reshape(dst.shape[0]*dst.shape[1],).tolist()



        dict_freq = {}
        for digit in range(1,10):
            dict_freq[str(digit)] = 0

        for value in col:
            if (type(value) is int or type(value) is float) and value>0:
                f_digit = firstDigit(value*1000000000000000000000000)
                dict_freq[str(f_digit)]+=1

        benford_dict = {}
        for digit in range(1,10):
            benford_dict[str(digit)] = math.log10(1+1/digit)
        freq_benf = np.array(list(benford_dict.values()))
        total_entries = 0

        for i in list(dict_freq.values()):
            total_entries+=i
        digits = list(dict_freq.keys())

        frequency = list(dict_freq.values())

        benford_freq = freq_benf * total_entries

        differences = np.subtract(frequency,benford_freq)
        abs_differences = abs(differences)
        sum_abs_differences = sum(abs_differences)

        percentage_difference_to_benford = (sum_abs_differences / sum(benford_freq)) * 100
        return round(percentage_difference_to_benford,2)
    except:
        print("Bild konnte nicht analysiert werden")
        return None
```

```python
directorys = next(os.walk('../airBnbPhotosOfRyanairDestinations/'))[1]
df2 = pd.DataFrame()
for directory in directorys:
    percentages = list()
    files = sorted(os.listdir('../airBnbPhotosOfRyanairDestinations/'+directory))
    for file in files :
    
        filename = os.fsdecode(file)
        if filename.endswith("jpg"):
            if filename not in df2.columns:
               df2[filename] = None
            percentages.append(graph('../airBnbPhotosOfRyanairDestinations/'+directory+'/'+filename,0,directory,filename))
            
            if file == files[-1]:
                len_diffenerce = abs(len(percentages)-len(df2.columns))
                if (len_diffenerce != 0) :
                    percentages.extend([None] * len_diffenerce)
                df2.loc[directory] = percentages


df2.to_excel("output_airbnb.xlsx")
```
